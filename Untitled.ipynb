```python
import random

d = []
for i in range(100): 
    o = []
    for j in range(12):
        o.append(random.random())
    d.append(o)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
d2 = scale(d, axis=1)#, with_mean=False, with_std=False)
```

```python
kk = KMeans(8).fit_predict(d)
```

```python
from matplotlib import pyplot as p
for i in range(len(kk)):
    if kk[i] == 0:
        p.plot(d[i])
p.show()
```
